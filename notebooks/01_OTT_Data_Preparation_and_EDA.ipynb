```python
dataset_02 = dataset.copy()
```

```python
dataset_02.rename(columns={"type": "content_type"},inplace=True)
```

```python
dataset_02["director"]=dataset_02["director"].fillna("Unknown")
```

```python
dataset_02["cast"] = dataset_02["cast"].fillna("Unknown")
dataset_02["country"] = dataset_02["country"].fillna("Unknown")
dataset_02["rating"] = dataset_02["rating"].fillna("Unknown")
dataset_02["duration"] = dataset_02["duration"].fillna("Unknown")
dataset_02["description"] = dataset_02["description"].fillna("No Description")
```

```python
dataset_02.isnull().sum()
```

```python
dataset_02["primary_genre"] = dataset_02["genre"].str.split(",").str[0]
```

```python
dataset_02.to_csv(r"C:\Data analysis project\OTT-Audience-Intelligence-Dashboard\datasets\cleaned_dataset.csv",index=False)
```

```python
sorted(dataset_02["primary_genre"].unique())
```

```python
dataset_02["primary_genre"] = dataset_02["primary_genre"].replace({
    "Dramas": "Drama",
    "Comedies": "Comedy",
    "Documentaries": "Documentary",
    "Thrillers": "Thriller",
    "Action-Adventure": "Action & Adventure"
})
```
