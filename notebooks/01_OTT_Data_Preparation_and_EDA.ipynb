1. Amazon Prime has the largest content library, contributing 42% of all titles in the dataset.

2. Netflix is the second-largest platform with 38.3% of total content.

3. Movies dominate OTT platforms, accounting for significantly more titles than TV Shows.

4. Hulu is the only platform where TV Shows slightly exceed Movies, indicating a stronger focus on episodic content.

5. Drama is the most popular genre overall and is the leading genre on both Amazon Prime and Netflix.

6. Disney+ primarily focuses on Action & Adventure content, while Hulu's dominant genre is Action.

7. The United States is the largest content-producing country, followed by India and the United Kingdom.

8. TV-MA and TV-14 are the most common content ratings across OTT platforms.

9. OTT content production increased rapidly after 2015, showing significant industry growth.

10. The years 2019, 2020, and 2021 recorded the highest number of content releases, marking the peak period of OTT expansion.
